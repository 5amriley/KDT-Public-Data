```python
import pandas as pd
import numpy as np
```

```python
filename = './Table.csv'
```

```python
us_wageDF = pd.read_csv(filename, skiprows=3, header=0, index_col = 0, skipfooter=11, engine='python')
us_wageDF
```

```python
# 컬럼명 변경
us_wageDF = us_wageDF.rename(columns={'Unnamed: 1':'Category'})
us_wageDF
```

```python
# 인덱스를 RangeIndex로 설정
us_wageDF = us_wageDF.reset_index(drop=True)
us_wageDF
```

```python
indent = []
for idx in range(len(us_wageDF)):
    fst = us_wageDF.iloc[idx, 0].strip()[0]
    indent.append(us_wageDF.iloc[idx, 0].find(fst))
indent[:10]
```

```python
indent.count(8)
```

```python
first_mask = [True if indent[i] == 8 else False for i, _ in enumerate(indent)]
us_wageDF1 = us_wageDF[first_mask]
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

years = us_wageDF1.columns[1:]
indices = us_wageDF1.sort_values(by='2022', ascending=False)[:10].index
labels = {51:'정보', 68:'회계, 재무', 56:'금융 / 보험', 10:'전력 / 천연가스', 64:'법률 / 과학 등 전문직', 
          6:'석유 / 가스 추출', 34:'도매상', 86:'연방 공무원 (군인 등)', 12:'제조업',61:'부동산 및 임대업'}
colorlist = ['#55BEEB', '#97010E', '#C7E954', 'linen', '#99864D', 
             '#B8CABA', 'darkturquoise', 'oldlace', 'lavenderblush', 'cornflowerblue']
linestyles = ['-', ':', ':', ':', '-', ':', '-', ':', ':', ':']
plt.figure(figsize=(30, 20))
plt.title('미국 업종별 정규직 근로자 1인당 임금 추이표 (대분류)', fontsize=50, pad=20)   # pad : 타이틀과 그래프와의 간격
i = 0
for idx in indices:
    plt.plot(years, us_wageDF.iloc[idx, 1:], label=labels[idx], linewidth=5, c=colorlist[i], linestyle=linestyles[i])
    i += 1
plt.grid(axis='y')    # y축만 격자 표시
plt.legend(fontsize=25)
plt.xticks(us_wageDF1.columns[1::2], fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
